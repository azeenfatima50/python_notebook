```python
import yfinance as yf
import os

# Get current working directory (same location as the notebook)
base_dir = os.getcwd()

# Download Tesla, Apple, Microsoft data
tickers = ["TSLA", "AAPL", "MSFT"]
data = yf.download(tickers, start="2018-01-01", end="2026-02-09")

# Save CSV in the same folder
file_path = os.path.join(base_dir, "stock_market_data.csv")
data.to_csv(file_path)

print("Dataset saved at:", file_path)
```
